```python
from ultralytics import YOLO

# 모델 로드
model = YOLO("yolov8n.pt")

# 이미지 추론
results = model("7-2. 갈변된 사과.jpg")
# 결과 가져오기
result = results[0]

# 박스 정보
boxes = result.boxes
names = result.names

# 좌표 출력
for box in boxes:
    # 좌표 (xmin, ymin, xmax, ymax)
    x1, y1, x2, y2 = box.xyxy[0].tolist()

    # confidence score
    conf = box.conf[0].item()

    # 클래스 id
    cls = int(box.cls[0].item())

    label = names[cls]
    print(f"{label} ({conf:.2f})")
    print(f"bbox: ({x1:.1f}, {y1:.1f}, {x2:.1f}, {y2:.1f})")
# 결과 출력
import matplotlib.pyplot as plt

img = result.plot()
plt.imshow(img)
plt.axis('off')
plt.show()
```
